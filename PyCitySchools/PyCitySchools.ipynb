```python
#dependencies
import pandas as pd
```

```python
#file path for given files
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
#read school and student data and store in dataframes
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
#merge the 2 dataframes into one 
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#calculation of total schools, total students, and total budget
total_schools =len(school_data_complete["school_name"].unique())
total_students = school_data_complete["Student ID"].count()
total_budget = school_data_complete["budget"].sum()

#calculation of average math and reading score 
average_math = school_data_complete["math_score"].mean()
average_reading = school_data_complete["reading_score"].mean()
```

```python
#calculation for percent math, reading, and overal passing
passing_math_percent = (school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"])/total_students*100
passing_reading_percent = (school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"])/total_students*100
passing_mathreading_percent = (school_data_complete[(school_data_complete["math_score"] >= 70)& (school_data_complete["reading_score"] >= 70)].count()["student_name"])/total_students*100
```

```python
#put all calulations in a dataframe and format them accordingly using .map
district_summary = pd.DataFrame({"Total Schools": [total_schools], "Total Students": [total_students], "Total Budget": total_budget, "Average Math Score": [average_math], "Average Reading Score": [average_reading], "% Passing Math": [passing_math_percent], "% Passing Reading": [passing_reading_percent], "% Overall Passing": [passing_mathreading_percent]})
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:,.2f}%".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:,.2f}%".format)
#display dataframe
district_summary
```

```python
#set index school name and have a column for school type
school_type=school_data.set_index(["school_name"])["type"]

#calculate number of students, budget, and budget per student in each school
school_student_count = school_data_complete["school_name"].value_counts()
school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_student_budget = school_budget/school_student_count

#average math and reading score for each school
average_math_school = school_data_complete.groupby(["school_name"]).mean()["math_score"]
average_reading_school = school_data_complete.groupby(["school_name"]).mean()["reading_score"]


```

```python
#students passing math, reading, and both
school_passing_math = school_data_complete[(school_data_complete["math_score"]>=70)]
school_passing_reading= school_data_complete[(school_data_complete["reading_score"]>=70)]
school_passing_both = school_data_complete[(school_data_complete["reading_score"]>=70) & (school_data_complete["math_score"]>=70)]
#calculate percent of students passing math, reading, and both
school_passing_math_percent = (school_passing_math.groupby(["school_name"]).count()["student_name"]/school_student_count)*100
school_passing_reading_percent = (school_passing_reading.groupby(["school_name"]).count()["student_name"]/school_student_count)*100
school_passing_both_percent = (school_passing_both.groupby(["school_name"]).count()["student_name"]/school_student_count)*100
```

```python
#put all calulations in dataframe and format them accordingly using .map
school_summary = pd.DataFrame({"School Type": school_type, "Total Students": school_student_count, "Total School Budget": school_budget, "Per Student Budget": per_student_budget, "Average Math Score": average_math_school, "Average Reading Score": average_reading_school, "% Passing Math": school_passing_math_percent, "% Passing Reading": school_passing_reading_percent, "% Overall Passing": school_passing_both_percent})
school_summary["Total Students"] = school_summary["Total Students"].map("{:,}".format)
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,.2f}".format)
school_summary["Average Math Score"] = school_summary["Average Math Score"].map("{:,.2f}".format)
school_summary["Average Reading Score"] = school_summary["Average Reading Score"].map("{:,.2f}".format)
school_summary["% Passing Math"] = school_summary["% Passing Math"].map("{:,.2f}%".format)
school_summary["% Passing Reading"] = school_summary["% Passing Reading"].map("{:,.2f}%".format)
school_summary["% Overall Passing"] = school_summary["% Overall Passing"].map("{:,.2f}%".format)
#display dataframe
school_summary
```

```python
#put % overall passing from highest to lowest
top_performing_schools = school_summary.sort_values(["% Overall Passing"], ascending=False)
#display top 5 of top performing schools
top_performing_schools.head()
```

```python
#put % overall passing from lowest to highest
bottom_performing_schools = school_summary.sort_values(["% Overall Passing"], ascending=True)
#display top 5 of bottom performing schools
bottom_performing_schools.head()
```

```python
#seperate data into only ninth, tenth, seleventh, and twelfth grades
ninth_grade = school_data_complete[school_data_complete["grade"] == "9th"]
tenth_grade = school_data_complete[school_data_complete["grade"] == "10th"]
eleventh_grade = school_data_complete[school_data_complete["grade"] == "11th"]
twelfth_grade = school_data_complete[school_data_complete["grade"] == "12th"]

#mean math scores for each grade
ninth_grade_math = ninth_grade.groupby(["school_name"]).mean()["math_score"]
tenth_grade_math = tenth_grade.groupby(["school_name"]).mean()["math_score"]
eleventh_grade_math = eleventh_grade.groupby(["school_name"]).mean()["math_score"]
twelfth_grade_math = twelfth_grade.groupby(["school_name"]).mean()["math_score"]
#put values in data frame
math_scores_by_grade = pd.DataFrame({"9th Grade": ninth_grade_math, "10th": tenth_grade_math, "11th": eleventh_grade_math, "12th": twelfth_grade_math})
#display dataframe
math_scores_by_grade
```

```python
#mean math scores for each grade 
ninth_grade_reading = ninth_grade.groupby(["school_name"]).mean()["reading_score"]
tenth_grade_reading = tenth_grade.groupby(["school_name"]).mean()["reading_score"]
eleventh_grade_reading = eleventh_grade.groupby(["school_name"]).mean()["reading_score"]
twelfth_grade_reading = twelfth_grade.groupby(["school_name"]).mean()["reading_score"]
#put values in data frame
reading_scores_by_grade = pd.DataFrame({"9th Grade": ninth_grade_reading, "10th": tenth_grade_reading, "11th": eleventh_grade_reading, "12th": twelfth_grade_reading})
#display dataframe
reading_scores_by_grade
```

```python
#bins and group names for spending amounts per student
spending_bins=[0,585,630,645,680]
group_names= ["<$585","$585-630","$630-645","$645-680"]

#copy data frame
school_spendingdf = school_summary

#use .cut to use the bin and group names to add another column in dataframe (Spending range per student)
school_spendingdf["Spending Ranges(Per Student)"] = pd.cut(per_student_budget, spending_bins, labels=group_names)
school_summary ["Spending Ranges(Per Student)"]= pd.cut(per_student_budget, spending_bins, labels=group_names)

#copy data frame
school_spending_copy=school_spendingdf
#change string to float values for columns in dataframe
school_spending_copy["Average Math Score"] = school_spending_copy["Average Math Score"].astype("float")
school_spending_copy["Average Reading Score"] = school_spending_copy["Average Reading Score"].astype("float")
school_spending_copy["% Passing Math"] = school_spending_copy["% Passing Math"].str.replace("%","").astype("float")
school_spending_copy["% Passing Reading"] = school_spending_copy["% Passing Reading"].str.replace("%","").astype("float")
school_spending_copy["% Overall Passing"] = school_spending_copy["% Overall Passing"].str.replace("%","").astype("float")
```

```python
#calculate the mean for math score, reading score, % math passing, %reading passing, and % overall passing based on spending range
spending_math_average = school_spending_copy.groupby(["Spending Ranges(Per Student)"]).mean()["Average Math Score"]
spending_reading_average = school_spending_copy.groupby(["Spending Ranges(Per Student)"]).mean()["Average Reading Score"]
spending_percentmath_average =school_spending_copy.groupby(["Spending Ranges(Per Student)"]).mean()["% Passing Math"]
spending_percentreading_average = school_spending_copy.groupby(["Spending Ranges(Per Student)"]).mean()["% Passing Reading"]
spending_percentoverall_average = school_spending_copy.groupby(["Spending Ranges(Per Student)"]).mean()["% Overall Passing"]
```

```python
#create dataframe and add calculations to dataframe and format them accordingly using .map
scores_by_school_spending = pd.DataFrame({"Average Math Score": spending_math_average, "Average Reading Score": spending_reading_average, "% Passing Math": spending_percentmath_average, "% Passing Reading": spending_percentreading_average, "% Overall Passing": spending_percentoverall_average})
scores_by_school_spending["Average Math Score"] = scores_by_school_spending["Average Math Score"].map("{:,.2f}".format)
scores_by_school_spending["Average Reading Score"] = scores_by_school_spending["Average Reading Score"].map("{:,.2f}".format)
scores_by_school_spending["% Passing Math"] = scores_by_school_spending["% Passing Math"].map("{:,.2f}%".format)
scores_by_school_spending["% Passing Reading"] = scores_by_school_spending["% Passing Reading"].map("{:,.2f}%".format)
scores_by_school_spending["% Overall Passing"] = scores_by_school_spending["% Overall Passing"].map("{:,.2f}%".format)
#display dataframe
scores_by_school_spending
```
